```python
import pandas as pd
df=pd.read_csv('weather_data.csv',parse_dates=['day'])
df.set_index('day',inplace=True)
```

```python
df
```

```python
new_df=df.fillna(0)
```

```python
new_df
```

```python
new_df1=df.fillna(df["temperature"].mean())
#Here what I did is that I replaced all the NaN values with the mean of the temperature which is not a good method.
```

```python
new_df1
```

```python
new_df=df.fillna(
{
    'temperature':df['temperature'].mean(),
    'windspeed':0,
    'event':'no event'
})
```

```python
new_df
```

Here, what I did is that i changed the missing value of windspeed as O, and the missing value in temperatue gets the mean temperature value, and event getting the no event value as we can see in the dataframe. 

Thus, that's how fillna method comes to apply.

Also, not to forget there is a method called ffill which uses the previous value of the row and it helps to give data an effective outlook whatsoever. 

```python
new_df=df.fillna(method="ffill")
```

```python
new_df
```

Here, you can see all the NaN values in the original dataframe gets the previous row value giving the data a different outlook, Right?

```python
new_df=df.fillna(method="bfill",axis="columns")
```

```python
new_df
```

```python
new_df=df.interpolate()
new_df
```

Here, the missing value gets filled by the value of the next column

```python
new_df=df.interpolate(method="time")
```
